# Zasady zaliczenia ćwiczeń

* Ćwiczenia z przedmiotu Sieci Neuronowe zalicza się na ocenę. 
* Na ćwiczeniach obowiązuje znajomość materiału z wykładu.
* Każda ocena poza **NZAL** pozwala na przystąpienie do egzaminu z przedmiotu.
* Ocena z ćwiczeń przyznawana jest na podstawie punktów zdobytych za **samodzielną** pracę nad zadaniami domowymi.


## Zadania Domowe

Na ćwiczeniach będą Państwo mieli do rozwiązania dwa rodzaje zadań domowych:

   1. Zadania Konkursowe - 3 zadania po 20 punktów każde - max. 60 punktów.
        * Polegają na rozwiązaniu konkretnego problemu z wyróżnionym minimalnym progiem dokładności (np. w formie minimalnej wartości *accuracy*). 
        * Ilość punktów zdobytych za zadanie będzie ustalona na podstawie rankingów z całego roku, przy czym za przekroczenie minimalnego progu z punktów wyżej automatycznie zdobywają państwo minimum 10 punktów.
        * Na każde zadania konkursowe będą mieli Państwo minimum 2 tygodnie na przesłanie rozwiązań.
        * Do każdego zadania oprócz specyfikacji warunków zadania (rodzaj problemu, dozwolone techniki itp.) będą Państwo mieli dostarczone dane wraz z kodem do ich wczytania oraz kod do testowania swoich rozwiązań.
        * Jako rozwiązanie zadania przesyłają Państwo zapisane wagi modelu wraz z kodem, którego użyli Państwo do nauczenia tego modelu.
         
         
   2. Zadania domowe - max. 6 zadań po 6 punktów - max. 36 punktów.
       * Polegają najczęściej na uzupełnieniu brakującego kodu, tak aby konkretne warunki zostały spełnione (np. lepsza dokładność niż podany baseline).
   
## Końcowa Ocena

Dodatkowo do zdobycia mają Państwo 4 punktów za aktywność, przyznawane wg uznania prowadzącego ćwiczenia.

Obowiązują następujące progi punktowe:
* [0, 50) punktów: 2.0
* [50, 60) punktów: 3.0
* [60, 70) punktów: 3.5
* [70, 80) punktów: 4.0
* [80, 90) punktów: 4.5
* [90, 100) punktów: 5.0
   
## Obecność

* Obecność na ćwiczeniach jest obowiązkowa.
* W ciągu semestru dopuszcza się maksymalnie **3 nieusprawiedliwione nieobecności**, przekroczenie tego limitu skutkuje oceną **NZAL** z ćwiczeń i niedopuszczeniem do egzaminu.

# Środowisko pracy

Na zajęciach będzie korzystać głównie z pakietu PyTorch>=1.0.0 dla Pythona w wersji 3.6. Na własną odpowiedzialność można korzystać z innych wersji Pythona. Poprzednie wersje PyTorcha są niezalecane.

## Lokalnie

1. Ściągnąć repozytorium przedmiotu: `git clone https://github.com/gmum/nn2019`  
(Można również ściągnąć zipa repozytorium "ręcznie")

2. Sciągnąć miniconde: https://conda.io/en/latest/miniconda.html (Niewymagane jeśli zainstalowana jest `Anaconda`)

3. Stworzyć środowisko razem z wymaganymi paczkami:  
`conda create --name nn2019 python=3.6 numpy=1.13.3 scipy=1.1.0 matplotlib tqdm scikit-learn=0.20.0 jupyter notebook`

4. Aktywować środowisko:  
Unix/MacOS: `source activate nn2019`  
Windows: w `Anaconda Prompt`: `activate nn2019`

5. Doinstalować PyTorcha: https://pytorch.org/  
Typowo: `conda install pytorch torchvision cudatoolkit=9.0 -c pytorch`

6. W folderze repo odpali serwer jupyter: `jupyter notebook`

7. Sprawdzić w przeglądarce czy wszystko działa
